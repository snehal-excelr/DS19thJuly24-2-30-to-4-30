```python
# from pandas import read_csv
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data
df = pd.read_csv('TSNE_data.csv')
```

```python
df.head() # M Malignant - Cancer Tumor present, B Benign - not a cancer tumor. We want to find if there is a pattern in M tumor and B tumor
```

```python
# Split-out validation dataset
array = df.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
array
```

```python
df
```

```python
#TSNE visualization
data_tsne = TSNE(n_components=2).fit_transform(X) # no of components = 2, max you can go for 3cluster.tsneplot(score=data_tsne)#visualize data
```

```python
data_tsne
```

```python
newdf=pd.DataFrame(data_tsne).join(df['diagnosis'])
newdf
```

```python
# get a list of categories
color_class = df['diagnosis'].to_numpy()
# Convert dataframe to array. legnedpos=position on upper right corner showing color difference, colorlist-show different color for M and B in diagnosis column
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15,1)) # legendanchor - sets position of M and B color box, if not given shows by default inside big rectangle
plt.show()
#Plot will be stored in the default directory
```
